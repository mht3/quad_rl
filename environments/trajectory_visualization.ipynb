```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
n_waypoints = 64
n = 2         # y oscillations
m = 4          # z oscillations
z_offset = 3.0
phi = np.pi / 4
psi = np.pi / 4

alpha = 2
beta = 2
gamma = 1.1
```

```python

t = np.linspace(0, 2*np.pi, n_waypoints)

x = alpha * np.sin(t)
y = beta * np.sin(n * t + phi)
z = gamma * np.sin(m * t + psi) + z_offset

fig = plt.figure(figsize=(6, 12))
ax = fig.add_subplot(111, projection='3d')

ax.plot(x, y, z, 'b:', label='Trajectory', linewidth=2)
# ax.plot(x, y, z, 'ko', alpha=0.3, markersize=3, label='Reference Pose')

# labels
ax.set_xlim(-2.5, 2.5)
ax.set_ylim(-2.5, 2.5)
ax.set_zlim(0, 5)
ax.set_xlabel('x (m)')
ax.set_ylabel('y (m)')
ax.set_zlabel('z (m)')
ax.legend()
ax.set_box_aspect([1,1,1])

plt.show()
```

```python
n_waypoints = 64
n = 2         # y oscillations
m = 4          # z oscillations
z_offset = 3.0
phi = np.pi / 4
psi = np.pi / 4

alpha = 2
beta = 2
gamma = 1.1
t = np.linspace(0, 2*np.pi, n_waypoints)

x = alpha * np.sin(t)
y = beta * np.sin(n * t + phi)
z = gamma * np.sin(m * t + psi) + z_offset

sigma = 0.2
x += np.random.normal(loc=0.0, scale=sigma, size=(n_waypoints,))
y += np.random.normal(loc=0.0, scale=sigma, size=(n_waypoints,))
z += np.random.normal(loc=0.0, scale=sigma, size=(n_waypoints,))
fig = plt.figure(figsize=(6, 12))
ax = fig.add_subplot(111, projection='3d')

ax.plot(x, y, z, 'b:', label='Trajectory', linewidth=2)

# labels
ax.set_xlim(-2.5, 2.5)
ax.set_ylim(-2.5, 2.5)
ax.set_zlim(0, 5)
ax.set_xlabel('x (m)')
ax.set_ylabel('y (m)')
ax.set_zlabel('z (m)')
ax.legend()
ax.set_box_aspect([1,1,1])

plt.show()
```
